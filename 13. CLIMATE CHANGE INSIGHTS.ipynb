## EDA (Exploratory Data Analysis)

# Box Plot

```python
# Box Plot for Temperature by Country
plt.figure(figsize=())
sns.boxplot(x='Country', y='Temperature', data=df)
plt.title('Temperature Box Plot by Country')
plt.xticks(rotation=90)
plt.show()
```

# Count Plot

```python
# Count Plot for Location
plt.figure(figsize=(20,20))
sns.countplot(x='Location', data=df)
plt.title('Count of Locations')
plt.xticks(rotation=90)
plt.show()

# Count Plot for Country
plt.figure(figsize=(20,20))
sns.countplot(x='Country', data=df)
plt.title('Count of Countries')
plt.xticks(rotation=45)
plt.show()
```

# Scatter Plot

```python
# Scatter Plot for Temperature vs CO2 Emissions
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature', y='CO2 Emissions', data=df)
plt.title('Temperature vs CO2 Emissions')
plt.show()
```

# Autocorrelation Plot

```python
from pandas.plotting import autocorrelation_plot

# Autocorrelation Plot for Temperature
plt.figure(figsize=(10, 6))
autocorrelation_plot(df['Temperature'])
plt.title('Autocorrelation Plot of Temperature')
plt.show()
```

# Heatmap

```python
# Heatmap for correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df[['Temperature', 'CO2 Emissions', 'Sea Level Rise', 'Precipitation', 'Humidity', 'Wind Speed']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

# histplot

```python
# Histogram for Temperature
plt.figure(figsize=(10, 6))
sns.histplot(df['Temperature'], bins=30, kde=True)
plt.title('Distribution of Temperature')
plt.show()
```

# Violin Plot

```python
# Violin Plot for Temperature by Country
plt.figure(figsize=(12, 6))
sns.violinplot(x='Country', y='Temperature', data=df)
plt.title('Temperature Distribution by Country')
plt.xticks(rotation=90)
plt.show()
```

# Pair Plot

```python
# Pair Plot for numerical features
sns.pairplot(df[['Temperature', 'CO2 Emissions', 'Sea Level Rise', 'Precipitation', 'Humidity', 'Wind Speed']])
plt.title('Pair Plot of Numerical Features')
plt.show()
```

# Joint Plot

```python
# Joint Plot for Temperature and CO2 Emissions
sns.jointplot(x='Temperature', y='CO2 Emissions', data=df, kind='scatter')
plt.title('Joint Plot of Temperature and CO2 Emissions')
plt.show()
```
